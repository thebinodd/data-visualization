```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv("./Datasets/iris.csv")
iris.head(5)
```

```python
iris.sample(5) #There are 3 different species of Iris Flower on the basis of their length of its sepal and petal
```

```python
#Coloured Encoded Scattered Plots
#We have to plot the scatter between sepal and petal and show which point represents which species
#For that we need to assign unique index for each Species

iris["Species"] = iris["Species"].map({
    "Iris-versicolor":0,
    "Iris-virginica":1,
    "Iris-setosa":2
})
iris.sample(5)
```

```python
plt.scatter(iris["SepalLengthCm"] , iris["PetalLengthCm"] , 

            c=iris["Species"], #Gives unique colors for each species
            cmap = "winter", #Theme
            alpha = 0.9 #Opacity
           )
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.colorbar() #Gives the colorbar for better understanding
plt.show()
plt.savefig("./Images/coloredScatterPlot.png")
```
