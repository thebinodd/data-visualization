```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./Datasets/batter.csv")
df.head()
```

```python
sampled_df = df.head(100).sample(25 , random_state = 5)
#Random state acts as the seed for certain sample. For the same dataframe, everytime with random_state=5 gives the same sample
#If you use the same dataframe and and same random_state, you will get same sampled dataframe as below
sampled_df
```

```python
plt.scatter(sampled_df["avg"] ,sampled_df["strike_rate"] )
plt.show()
#Since every dot represents the unique batsman, it is difficult to know which dot represents which batsman
```

```python
x = [1,2,3,4]
y = [5,6,7,8]
plt.scatter(x,y)
#plt.text(x-coordinate , y-coordinate , "name")
plt.text(1,5 , "A")
plt.text(2,6 , "B")
plt.text(3,7 , "C")
plt.text(4,8 , "D")
plt.show()
```

```python
plt.figure(figsize=(18,10)) #Width and Height of Plot
plt.scatter(sampled_df["avg"] ,sampled_df["strike_rate"] , s=sampled_df["runs"] )
for i in range(sampled_df.shape[0]):
    plt.text(
        sampled_df["avg"].values[i],
        sampled_df["strike_rate"].values[i],
        sampled_df["batter"].values[i],
    )
plt.axhline(130 , color="red") #Give the horizontal line from 130
plt.axvline(30 , color="red") #Give the vertical line from 30
plt.show()
```
