```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
batter = pd.read_csv("./Datasets/batter.csv")
batter.sample(5)
```

```python
plt.style.available
plt.style.use("fivethirtyeight")
```

```python
#One way to plot scatter the plots between average and strike rate


#For Runs vs Strike Rate
plt.figure(figsize=(8,4))
plt.scatter(batter["runs"] , batter["strike_rate"])
plt.xlabel("Runs")
plt.ylabel("Strike Rate")
plt.show()

#For Runs Vs Average
plt.figure(figsize=(8,4))
plt.scatter(batter["runs"] , batter["avg"])
plt.xlabel("Runs")
plt.ylabel("Average")
plt.show()

#We need two separate code for two plots and we get two different figures
```

```python
#Another Way to Plot Same Graphs

fig , ax = plt.subplots(nrows=2, ncols=1 , sharex=True ,) 
#The function subplots returns two things , one is figure and another is axes. 
#Stored those values in variables named fig and ax
#The nrows = 2 and ncols=1 means we are making 2*1 = 2 plots in single figure
#ax is the array containing two axes where ax[0] is one plot where ax[1] is another

#For Upper Graph
ax[0].scatter(batter["runs"] , batter["avg"] ,  color="blue" , marker="*")
ax[0].set_title("Runs vs Average")
ax[0].set_ylabel("Average")

#For Lower Graph
ax[1].scatter(batter["runs"] , batter["strike_rate"] , marker="+" , color="red")
ax[1].set_title("Runs vs Strike Rate")
ax[1].set_ylabel("Strike Rate")
ax[1].set_xlabel("Runs") #This is same for both graphs

```

```python
#Figure with 4 plots in single image
fig , ax = plt.subplots(nrows=2, ncols=2 , figsize=(15,6))
#For Upper Left Graph
ax[0][0].scatter(batter["runs"] , batter["avg"] ,  color="blue" , marker="*")
ax[0][0].set_title("Runs vs Average")
ax[0][0].set_ylabel("Average")
ax[0][0].set_xlabel("Runs") 

#For Upper Right Graph
ax[0][1].scatter(batter["runs"] , batter["strike_rate"] , marker="+" , color="red")
ax[0][1].set_title("Runs vs Strike Rate")
ax[0][1].set_ylabel("Strike Rate")
ax[0][1].set_xlabel("Runs") 

#For Bottom Left

ax[1][0].hist(batter["avg"] , bins=[5,10,15,20,25,30,35,40,45,50,55,60] , color="green" )
ax[1][0].set_xlabel("Average")
ax[1][0].set_ylabel("No. of Batsman") 

#For Bottom Right
custom_bins = np.arange(20,400,10).tolist()
ax[1][1].hist(batter["strike_rate"] , bins=custom_bins , color="pink" )
ax[1][1].set_xlabel("Strike Rate")
ax[1][1].set_ylabel("No. of Batsman")
#ax[1][1].axvline(130) for custom line in plot
```
