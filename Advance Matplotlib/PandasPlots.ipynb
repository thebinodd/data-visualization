```python
import pandas as pd
import seaborn as sns
```

```python
# pd.plot() is used to plot all the graphs of matplotlib in pandas
# low effort and low customization
# Works on both series and dataframe
```

```python
s = pd.Series([1,2,3,4,5,6])
s.plot(kind="line")
```

```python
tips = sns.load_dataset("tips") #Inbuilt dataset on seaborn
tips.head()
```

```python
tips.plot(
    kind="scatter",
    x="total_bill",
    y="tip",
    title="Cost Analysis",
    marker="+",
    c="sex" ,  #Different colors for different sexes
    cmap="viridis"
)
```

```python
stocks = pd.read_csv("https://raw.githubusercontent.com/m-mehdi/pandas_tutorials/main/weekly_stocks.csv")
stocks.head()
```

```python
stocks["FB"].plot(kind="line") #Line plot for series
```

```python
stocks.plot(kind="line") #For DataFrame
```
