```python
tips = sns.load_dataset("tips") #Inbuilt dataset on seaborn
tips.head()
```

```python
tips.plot(
    kind="scatter",
    x="total_bill",
    y="tip",
    title="Cost Analysis",
    marker="+",
    c="sex" ,  #Different colors for different sexes
    cmap="viridis"
)
```

```python
stocks = pd.read_csv("https://raw.githubusercontent.com/m-mehdi/pandas_tutorials/main/weekly_stocks.csv")
stocks.head()
```

```python
stocks["FB"].plot(kind="line") #Line plot for series
```

```python
stocks.plot(kind="line") #For DataFrame
```

```python
stocks[["MSFT"  , "FB" , "Date"]].plot(kind="line" , x="Date") #For only two series
```

```python
#Single BarCharts
temp = pd.read_csv("./Datasets/batsman_season_record.csv").head()
temp
# temp.plot(kind="bar") #For DataFrames
temp["2015"].plot(kind="bar") #For Series
```

```python
#For Multiple Barcharts
temp.plot(kind="bar" , subplots=True , figsize=(10,8))
```

```python
#For Stacked Barcharts
temp.plot(kind="bar" , stacked=True)
```

```python
#histograms
# stocks.plot(kind="hist") #For Dataframes
stocks["FB"].plot(kind="hist" , bins=30)
```

```python
#Single PieChart
df = pd.DataFrame(
    {
        'batsman':['Dhawan','Rohit','Kohli','SKY','Pandya','Pant'],
        'match1':[120,90,35,45,12,10],
        'match2':[0,1,123,130,34,45],
        'match3':[50,24,145,45,10,90]
    }
)

df["match1"].plot(kind="pie" , labels=df["batsman"].values , figsize=(15,6))
```

```python
#Multiple Pie Charts

df[["match1" , "match2" , "match3"]].plot(kind="pie" , subplots=True , figsize=(15,6) , labels=df["batsman"].values , autopct = "%0.1f%%")
```
