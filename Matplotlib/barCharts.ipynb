```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
children = [10 , 30 , 15 , 35 , 25]
colors = ["Red" , "Blue", "Aqua" , "Pink" , "Green"]
plt.bar(colors , children)
```

```python
plt.barh(colors , children) #For Horizontal Bars
```

```python
df = pd.read_csv("./Datasets/batsman_season_record.csv")
df
```

```python
#For Individual Year
plt.bar(df["batsman"] , df["2016"])
```

```python
#For all years side by side 

#for 2015
plt.bar(np.arange(df.shape[0])-0.3 , df["2015"] , width=0.3) #shift this bar left side

#for 2016
plt.bar(np.arange(df.shape[0]) , df["2016"] , width = 0.3)

#for 2017
plt.bar(np.arange(df.shape[0]) + 0.3 , df["2017"] , width = 0.3) #shift this bar right side

#These three lines of code makes three barchart allined in one barchart area

#for ranaming x values
plt.xticks(np.arange(df.shape[0]) , df["batsman"])

#for clear show
plt.show()
```
