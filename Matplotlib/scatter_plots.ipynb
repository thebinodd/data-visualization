```python
df = pd.read_csv("./datasets/batter.csv").head(50)
```

```python
avg = df["avg"] #Average runs of batsman in match
st_rate = df["strike_rate"] #Strike of batsman
```

```python
plt.scatter(avg , st_rate , color="red") 
plt.title("Average vs Strike Rate of Batsman in IPL")
plt.xlabel("Average")
plt.ylabel("Strike Rate")
# bottom left dot represents low average low strike rate
#top right dots represent high average high strike rate
```
