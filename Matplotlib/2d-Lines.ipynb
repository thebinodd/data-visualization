```python
batsman = pd.read_csv("Datasets/sharma-kohli.csv")
```

```python
plt.plot(batsman["index"] , batsman["V Kohli"]) #Show only for V Kohli
plt.plot(batsman["index"] , batsman["RG Sharma"]) #Show for both BAtsman
plt.title("Virat Kohli vs Rohit Sharma Run Comparison")
plt.xlabel("Season")
plt.ylabel("Runs")
```

```python
plt.plot(batsman["index"] , batsman["V Kohli"] , 
         color="red",
         linestyle="dashed",
         linewidth=3,
         marker="D",
         markersize=7,
)
plt.plot(batsman["index"] , batsman["RG Sharma"],
        color="blue",
        linestyle="dashdot",
        linewidth = 2,
        marker="+",
        markersize=9
) 
plt.title("Virat Kohli vs Rohit Sharma Run Comparison")
plt.xlabel("Season") #Labeling the X axis
plt.ylabel("Runs")
plt.grid() #Give grid for better value reading

plt.xlim(2010 , 2014) #Zoom in for the values of 2010 to 2014 in x-axis
plt.ylim(200 , 700)

plt.show() #This function is used to show graph in other ide and libraries like flask
```
