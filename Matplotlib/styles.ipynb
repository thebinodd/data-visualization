```python
import matplotlib.pyplot as plt
```

```python
#2d - plots
data = [20 , 10 , 40 , 30 ,55 , 20]
sub = ["MTH" , "PHY" , "BIO" , "ENG" , "CHEM" , "NEP"]
plt.plot(sub ,data)

#Default Styling
```

```python
plt.style.available
```

```python
plt.style.use("fivethirtyeight") #New Style
data = [20 , 10 , 40 , 30 ,55 , 20]
sub = ["MTH" , "PHY" , "BIO" , "ENG" , "CHEM" , "NEP"]
plt.plot(sub ,data)
```

```python
#Bar Graph
plt.style.use("ggplot")
data = [20 , 10 , 40 , 30 ,55 , 20]
sub = ["MTH" , "PHY" , "BIO" , "ENG" , "CHEM" , "NEP"]
plt.bar(sub ,data)

plt.savefig("Barchart.png") #Save as Image of given plot
```

```python
plt.style.use("dark_background")
data = [20 , 10 , 40 , 30 ,55 , 20]
sub = ["MTH" , "PHY" , "BIO" , "ENG" , "CHEM" , "NEP"]
plt.pie(data ,labels=sub)
plt.show()
```
